## Tratamento de dados faltantes

```python
# importando os pacotes necessarios 

import pandas as pd
```

```python
# importando e imprimindo a base de dados 
dados =  pd.read_csv('dados/aluguel_residencial_01.csv', sep = ';')
dados.head(10)
```

```python
# metodos para visualizar os dados nulos 
dados.isnull()
```

```python
# metodos para visualizar os dados nulos 
dados.notnull()
```

```python
# metodos para visualizar os dados nulos 
dados.info()
```

```python
# Vendo os valores nulos no campo valor 

dados[dados['Valor'].isnull()]
```

```python
# Eliminandos valores nulos de valor, pois não fazem sentido

A = dados.shape[0]
dados.dropna(subset = ['Valor'], inplace = True)
B = dados.shape[0]
A - B
```

```python
# Vendo os valores nulos no campo valor  após o tratamento

dados[dados['Valor'].isnull()]
```

```python
# Avaliando quantidade de valores nulos em Condominio

dados[dados['Condominio'].isnull()].shape[0]
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
# Eliminandos valores nulos de condominio QUE não fazem sentido

A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
# Contando os valores nulos
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# Subistituindo os valores NULOS por ZERO onde faça sentido

dados.fillna(0, inplace = True)
```

```python
# Subistituindo os valores NULOS por ZERO onde faça sentido [ Outra forma de fazer a operação de cima, incluindo IPTU]
dados  = dados.fillna({'Condominio':0, 'IPTU':0})
```

```python
# Contando os valores nulos em condominio
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# Contando os valores nulos em IPTU
dados[dados['IPTU'].isnull()].shape[0]
```

```python
# Verificando o tratamento realizado

dados.info()
```

```python
dados.to_csv('dados/aluguel_residencial_02.csv', sep = ';', index = False)
```
