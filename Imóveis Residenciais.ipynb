# Relatorio de Análise II

## Imóveis Residenciais

```python
# importando os pacotes necessarios

import pandas as pd
```

```python
# importanto os dados para analise

dados =  pd.read_csv('dados/aluguel.csv', sep = ';')
```

```python
dados.head()
```

```python
# Criando uma lista recomendo os dados duplicados do campo TIPO 

list(dados['Tipo'].drop_duplicates())
```

```python
# armazenando o retorno dos tipos unicos em uma LISTA 

residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila'
              ]

residencial
```

```python
dados.head(5)
```

```python
# Itens que estão na lista residencial e no dataframe serão marcados como TRUE os que não pertencem FALSE
# Apos isso o retorno sera salvo em SELECAO

selecao = dados['Tipo'].isin(residencial)
selecao
```

```python
# Salvado os retornos TRUE em um novo dataframe

dados_residenciais =  dados[selecao]
dados_residenciais
```

```python
# Verificando se o metodo aplicado resultou na eliminação dos registros indesejados
list(dados_residenciais['Tipo'].drop_duplicates())
```

```python
# Verificando o comprimento do dataframe

dados_residenciais.shape[0]
```

```python
# Arrumando o index após os ajustes no dataset

dados_residenciais.index = range(dados_residenciais.shape[0])
dados_residenciais
```

## Exportando a base de dados reduzida em CSV

```python
# Comando para exportar os dados como CSV separados por ponto e virgula , colando index = FALSE para não exportar o index
dados_residenciais.to_csv('dados/aluguel_residencial_1.csv', sep = ';', index = False)
```
