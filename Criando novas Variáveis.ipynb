## Criando novas Variáveis

```python
# importando pacote utilizados
import pandas as pd
```

```python
# importando os dados trabalhados na ultima etapa
dados = pd.read_csv('dados/aluguel_residencial_02.csv', sep = ';')
```

```python
# visualizando as 10 primeiras linhas
dados.head(10)
```

```python
# calculando o valor bruto

dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados
```

```python
# Calculando valor do m² (por valor)

dados['Valor m2'] = (dados['Valor'] / dados['Area']).round(2)
```

```python
# Visualizando a coluna incluida
dados.head(10)
```

```python
# Calculando valor do m² (por valor bruto)

dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
# Visualizando a coluna incluida
dados
```

```python
# Incluindo uma nova variavel com a função lambda

casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']

dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head(10)
```

## Excluindo Variaveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
dados_aux
```

```python
# excluindo variavel de um DataFrame

del dados_aux['Valor Bruto']
```

```python
dados_aux
```

```python
# excluindo variavel de um DataFrame metodo 2
dados_aux.pop('Valor Bruto m2')
```

```python
dados_aux
```

```python
# excluindo variavel de um DataFrame metodo 3
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
dados.head(10)
```

```python
dados.to_csv('dados/aluguel_residencial_03.csv', sep = ';', index = False)
```
